Berikut adalah informasi atas penjelasan lengkap mengenai 11 kolom data yang ada di dalam variabel DataFrame yang bernama "data_patient_treatment" diantaranya adalah sebagai berikut:

8. MCV: Kolom yang berisikan data terkait hasil uji lab pasien mengenai MCV atau Mean Corpuscular Volume. MCV merupakan ukuran rata-rata dari sel darah merah milik pasien.
9. AGE: Kolom yang berisikan data mengenai umur dari setiap pasien.
10. SEX: Kolom yang berisikan data mengenai jenis kelamin pasien yang ditandai dengan nilai F (female/perempuan) dan M (male/laki-laki).
11. SOURCE: Terakhir adalah kolom yang berisikan data dari masing-masing status pasien apakah pasien tersebut merupakan pasien rawat inap (ditandai dengan data "in") atau pasien rawat jalan (ditulis/ditandai dengan data "out").
```python
"""
Menampilkan nilai rata-rata dari kadar kandungan leukosit/sel darah putih yang dimiliki oleh para pasien yang berada
di kisaran umur orang dewasa normal yaitu 19 - 65 tahun. Dengan kata lain, kita akan menampilkan rata-rata kadar
kandungan sel darah putih dari yang terendah yang dialami oleh para pasien dewasa rumah sakit
"""
mean_leucocyte = filtering_age_for_adult_patient.groupby("AGE")["LEUCOCYTE"].mean().sort_values(ascending = True)

print(mean_leucocyte)
```

```python
"""
Menampilkan perbandingan data antar status final yang memutuskan apakah pasien harus dirawat inap/rawat jalan
berdasarkan jumlah kadar kandungan sel darah putih (leukosit)
"""
plt.figure(figsize=(15, 8))

sns.barplot(x="Status", y="Leucocyte", hue="Status", data=data_patient_treatment)
plt.suptitle("Perbandingan Status Final Decision untuk Pasien", fontsize=15)
plt.title("Status Final Decision ini Bertujuan untuk Memutuskan Apakah Pasien \n"
          "Harus Tetap Dirawat Inap/Rawat Jalan berdasarkan Jumlah Leukosit", fontsize=13, pad=13, loc="center")
plt.xlabel("Final Decision", fontsize=12)
plt.ylabel("Jumlah Kadar Leukosit", fontsize=12)
#plt.savefig("../Data Visualization/Perbandingan Status Pasien Rawat Inap & Rawat Jalan berdasarkan Kadar Leukosit.png")
plt.show()
```

```python
# Melihat nilai korelasi/keterkaitan antar masing-masing data dari variabel DataFrame "data_patient_treatment"
plt.figure(figsize=(15, 8))

sns.heatmap(data_patient_treatment.corr(), cmap="Spectral", annot=True)
plt.suptitle("Korelasi/Keterkaitan Data di Masing-Masing Kolom", fontsize=15)
plt.title("Hubungan Timbal-Balik antara Kolom Data dalam bentuk Nilai Korelasi", fontsize=13, pad=13, loc="center")
#plt.savefig("../Data Visualization/Grafik Korelasi untuk Melihat Hubungan Sebab-Akibat antar Kolom.png")
plt.show()
```

Sebenarnya grafik heatmap diatas (atau yang biasa dikenal sebagai grafik peta panas) tujuannya adalah sama dengan fungsi ".corr()" yang dimana bertujuan untuk menampilkan nilai korelasi atau nilai hubungan timbal-balik antar kolom data dimana data tersebut bernilai numerik. Dengan mengetahui nilai korelasi yang ditunjukkan oleh grafik heatmap, kita dapat mengetahui apakah suatu kolom dengan kolom yang lain memiliki hubungan timbal-balik/keterhubungan yang sangat kuat atau tidak. 

Lalu untuk mengetahui informasi yang terkandung di dalam grafik heatmap kita harus mengetahui cara membaca nilai-nilai yang ditampilkan agar mendapatkan informasi yang ada di dalamnya dan mengambil kesimpulan apakah nilai data yang ada di dalam kolom tersebut saling memiliki keterkaitan atau tidak. Cara membaca grafik ini yaitu membaca secara vertikal/horizontal dengan memperhatikan nilai yang ada di dalam kotak sebagai penanda nilai korelasi. Berikut adalah penjelasan terkait nilai korelasi antar kolom yang ada di dalam variabel DataFrame "data_patient_treatment" setelah dilakukan proses preprocessing:

```python
"""
Menampilkan jumlah total persentase data dengan pie chart berdasarkan jumlah protein yang terkandung
di dalam Eritrosit (sel darah merah) antar jenis kelamin pasien
"""
plt.figure(figsize=(15, 8))

percentage_label = data_patient_treatment.groupby("Sex")["Hemoglobin"].sum().reset_index()
plt.pie(percentage_label["Hemoglobin"], labels = percentage_label["Sex"], colors=["cyan", "yellow"], autopct = "%1.1f%%")
plt.suptitle("Perbandingan Komposisi antar Jenis Kelamin Pasien", fontsize=15)
plt.title("Nilai Persentase dari Jumlah Total \n Hemoglobin per masing-masing Jenis Kelamin Pasien",
          fontsize=13, pad=13, loc="center")
plt.legend()
#plt.savefig("../Data Visualization/Persentase Pasien berdasarkan Jenis Kelamin dari Jumlah Total Hemoglobin.png")
plt.show()
```

Insight yang dapat kita ambil dari pie chart diatas adalah mengetahui komposisi/pembagian nilai data yang ada di dalam kolom "Sex" berdasarkan jumlah total hemoglobin yang dimiliki oleh masing-masing pasien yaitu Laki-Laki (nilai 1) dan Perempuan (nilai 0). Sebagai catatan bahwasanya jika kita menggunakan pie chart maka nilai persentasenya harus memenuhi 100%, tidak boleh kurang dari 100% atau melebihi 100% dan nilai 100% tersebut merupakan persentase dari total keseluruhan antar label data yang ada di dalam pie chart. Pada pie chart jika ditotalkan antara nilai 54.4 dengan 45.6 akan menghasilkan nilai yaitu 100. Selain itu, label yang dimiliki oleh setiap pie chart tidak boleh melebihi 6 karena jika suatu pie chart memiliki banyak sekali label akan menyulitkan membaca grafik tersebut apalagi mendapatkan insight yang ada di dalamnya. Maka dari itu pie chart diatas sudah memenuhi standar kriteria dalam membuat suatu grafik pie karena label yang dimiliki hanya 2 (nilai 0 untuk Jenis Kelamin Wanita dan nilai 1 untuk Jenis Kelamin Pria) sehingga kita akan mudah dalam mengambil wawasan yang ada di dalamnya serta total persentase keseluruhannya adalah 100%.

Pie chart diatas menunjukkan bahwasanya komposisi pasien yang memiliki jenis kelamin Pria memiliki komposisi terbanyak daripada pasien yang berjenis kelamin Wanita dengan jumlah persentase yaitu 54.4% sedangkan komposisi dari pasien berjenis kelamin Wanita hanya memiliki jumlah persentase yakni 45.6% saja. Jumlah persentase tersebut dilihat berdasarkan dari jumlah total Hemoglobin (protein yang terkandung di dalam Sel Darah Merah/Eritrosit) antar masing-masing jenis kelamin pasien.
