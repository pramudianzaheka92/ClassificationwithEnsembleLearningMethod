Tahap preprocessing selanjutnya adalah melakukan pengecekan apakah di dalam variabel DataFrame copy (new_df_patient) memiliki nilai data outliers atau tidak. Nilai outliers merupakan suatu nilai ekstrim di dalam kumpulan data. Alasan mengapa disebut sebagai nilai ekstrim adalah karena outliers memiliki nilai yang sangat jauh berbeda dari nilai mayoritas yang ada di dalam kumpulan data. Sebagai contoh pada kolom Grade (nilai) terdapat kumpulan nilai yang diantaranya adalah 80, 70, 76, 83, 60, 65, 55, 47, 43, 59 dan 400. Nilai 400 tersebutlah dapat dikatakan sebagai nilai outliers yang ada di dalam kolom Grade. Alasan kenapa nilai outliers harus dihilangkan dari kumpulan nilai data agar analisis statistik dan Algoritma Machine Learning yang di-build tidak menghasilkan hasil akhir berupa kesimpulan/analisis yang tidak tepat karena nilai outliers dapat memengaruhi perhitungan seperti rata-rata (mean) dan standar deviasi. Selain itu dengan dibiarkan nilai outliers di dalam kumpulan nilai data dapat membelokkan pengujian hipotesis yang sedang kita bangun dan alhasil hipotesis yang kita anggap tepat justru jauh dari kata tepat.

Tetapi sebelum menghapus nilai outliers di dalam "new_df_patient", saya akan melakukan pengecekan nilai outliers di beberapa kolom seperti "Hematocrit", "Hemoglobin", "Erythrocyte", "Leucocyte", "Thrombocyte", "Mch", "Mchc" dan "Mcv". Untuk mengecek apakah kolom-kolom tersebut terdapat nilai outliers/tidak saya akan menggunakan cara dengan memanfaatkan data visualisasi berupa grafik boxplot.
